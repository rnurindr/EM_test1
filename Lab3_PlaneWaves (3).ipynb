# Understanding plane waves
In this lab exercise, we are going to look into how plane waves in both frequency domain and time domain behave in a conductive Earth medium. <br>

After finishing this exercise, you will understand <br>
** for a plane wave in frequency domain**, 
* how the conductivity of the medium affects the skin depth; <br>
* how the frequency of the EM wave affects the skin depth; <br>
* how the plane wave behaves in spatial and time domain; <br>

** and for a plane wave in time domain**, 
* how the plane wave propogates in a conductive medium ; <br>
* how the conductivity affects the diffusion distance; <br>
* how the conductivity affects the speed at which a plane wave propogates; <br>
* how the peak velocity changes with time. <br>
